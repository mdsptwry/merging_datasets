```python
#get imports
import pandas as pd
import numpy as np
import matplotlib as plt
```

```python
# load datasets
disparity = pd.read_csv("cif_data/disparity.csv")
economics = pd.read_csv("cif_data/economics.csv")
environment = pd.read_csv("cif_data/environment.csv")
housing = pd.read_csv("cif_data/housing.csv")
sociodemo = pd.read_csv("cif_data/sociodemo.csv")
```

```python
merged_df = sociodemo.merge(economics, on='FIPS', how='inner', suffixes=('_disparity', '_economics'))\
               .merge(environment, on='FIPS', how='inner', suffixes=('', '_housing'))\
               .merge(housing, on='FIPS', how='inner', suffixes=('', '_environment'))\
               .merge(disparity, on='FIPS', how='inner', suffixes=('', '_sociodemo'))

merged_df.head(30)
```

```python
merged_df.to_csv('/Users/mdspatwary/merged_data.csv', index=False)
```
