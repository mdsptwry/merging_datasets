```python
#get imports
import pandas as pd
import numpy as np
import matplotlib as plt
```

Tract Datasets

```python
# load datasets
disparity = pd.read_csv("cif_data/disparity.csv")
economics = pd.read_csv("cif_data/economics.csv")
environment = pd.read_csv("cif_data/environment.csv")
housing = pd.read_csv("cif_data/housing.csv")
sociodemo = pd.read_csv("cif_data/sociodemo.csv")
```

```python
merged_df = sociodemo.merge(economics, on=['FIPS', 'Tract', 'County', 'State'], how='inner')\
               .merge(environment, on=['FIPS', 'Tract', 'County', 'State'], how='inner')\
               .merge(housing, on=['FIPS', 'Tract', 'County', 'State'], how='inner')\
               .merge(disparity, on=['FIPS', 'Tract', 'County', 'State'], how='inner')

merged_df.head(10)
```

County Dataset

```python
# load datasets
disparity_county = pd.read_csv("cif_data/disparity_county.csv")
economics_county = pd.read_csv("cif_data/economics_county.csv")
environment_county = pd.read_csv("cif_data/environment_county.csv")
housing_county = pd.read_csv("cif_data/housing_county.csv")
sociodemo_county = pd.read_csv("cif_data/sociodemo_county.csv")
```

```python
disparity_county.head()
```

Merging county datasets

```python
merged_county_df = sociodemo_county.merge(economics_county, on=['FIPS', 'County', 'State'], how='inner')\
               .merge(environment_county, on=['FIPS', 'County', 'State'], how='inner')\
               .merge(housing_county, on=['FIPS', 'County', 'State'], how='inner')\
               .merge(disparity_county, on=['FIPS', 'County', 'State'], how='inner')

merged_county_df.head()
```

I left out riskfactor to keep it consistent with our previous merged dataset.
